```python
# Importing the necessary tools
# General 
import pandas as pd
import numpy as np

# Web sraping
#!pip install beautifulsoup4
from bs4 import BeautifulSoup
import requests

# Regular Expressions
import re 
```

```python
names_df = pd.read_csv("/home/ec2-user/SageMaker/cheeses_names.csv")
```

```python
names_df.head()
```

```python
# Web scraping the cheeses names

cheese_name = [] # Creating an empty list for the cheeses names
for index, row in names_df.iterrows():
    query = "https://www.cheese.com/" + row["cheeses_names"] + "/"
    page = requests.get(query)
    soup = BeautifulSoup(page.content, 'html.parser')
    summary_title = soup.find('title')
    cheese_name.append(summary_title.text) # Saving the scraped data into the list
    print(summary_title.text)
```

```python
# Removing " - Cheese.com" from the each cheese name

name = []
for x in cheese_name:
    name.append(x.replace(" - Cheese.com", ""))
print(list(name))
```

```python
# Removing special characters

!pip install unidecode
import unidecode

updated_name_list = []

for item in name:
    updated_item = unidecode.unidecode(item)
    updated_name_list.append(updated_item)
print(updated_name_list)
```

```python
# Web scraping the cheeses characteristics

cheese_characteristics = [] # Creating an empty list for the cheeses characteristics
for index, row in names_df.iterrows():
    query = "https://www.cheese.com/" + row["cheeses_names"] + "/"
    page = requests.get(query)
    soup = BeautifulSoup(page.content, 'html.parser')
    summary_traits = soup.find('ul', class_="summary-points")
    cheese_characteristics.append(summary_traits.text) # Saving the scraped data into the list
    print(summary_traits.text)
```

```python
# Creating a dictionary with the cheese names and the cheese_characteristics

dictionary = dict(zip(updated_name_list, cheese_characteristics))
```

```python
# Creating a data frame from the dictionary

cheeses_df = pd.DataFrame([dictionary])
```

```python
# Transposing the index and the columns of the data frame

cheeses_df = cheeses_df.T
```

```python
# Showing the first five rows of the data frame

cheeses_df.head()
```

```python
# Using regular expressions to manipulate the data frame structure 

def create_attribute_column(df: pd.DataFrame, attribute: str):
    df[attribute] = [re.findall("(?<=" + attribute + ": )([^\\\n]+)", cell) for cell in df[0]]
    df[attribute] = df[attribute].apply(lambda x: ''.join(x))
    
def create_made_from_column(df: pd.DataFrame, attribute: str):
    df[attribute] = [re.findall("(?<=" + attribute + " )([^\\\n]+)", cell) for cell in df[0]]
    df[attribute] = df[attribute].apply(lambda x: ''.join(x))
    
```

```python
# Creating columns on the data frame for the attributes of each cheese

create_made_from_column(df=cheeses_df, attribute="Made from")

attribute_list = ["Country of origin", "Region", "Type", "Family", "Fat content", "Texture", "Rind", "Colour", "Flavour", "Aroma", "Vegetarian", "Producers"]

for item in attribute_list:
    create_attribute_column(df=cheeses_df, attribute=item)
    
```

```python
# Showing the first rows columns of the data frame

cheeses_df.head()
```

```python
# Removing the first column of the data frame
cheeses_df = cheeses_df.iloc[: , 1:]

cheeses_df.head()
```

```python
# Replacing empthy cells with "NaN"
cheeses_df  = cheeses_df.replace('', np.nan) 

cheeses_df.head()
```
